```python
import numpy as np
import matplotlib.pyplot as plt


rad = np.linspace(0, 5, 100)
azm = np.linspace(0, 2 * np.pi, 100)
r, th = np.meshgrid(rad, azm)
z = (r ** 2.0) / 4.0


fig = plt.figure()
plt.subplot(projection="polar")

plt.pcolormesh(th, r, z)

plt.plot(azm, r, color='k', ls='none') 
plt.grid()

plt.show()
```

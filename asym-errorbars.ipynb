```python
import numpy as np
from matplotlib import pyplot as plt

# generate data
x = np.linspace(0, 1, 5)
y = np.cos(x)

# generate errors
err = np.linspace(0.1, 0.25, 5)
lower_err = err * 0.5
upper_err = err 

# put lower and upper values it to the same list
asym_err = [lower_err, upper_err]

plt.scatter(x, y)
plt.errorbar(x, y, yerr=asym_err, xerr=asym_err)
```

```python
import matplotlib.pyplot as plt 
import numpy as np

x = np.arange(50)
xy = x*x[:,None]/200 # 0.000 to 12.005

plt.contour(xy, levels=12, colors='black').clabel(inline=1, fmt='%.2f')
c = plt.contourf(xy, levels=12)
plt.colorbar(c)
```
